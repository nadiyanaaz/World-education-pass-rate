```python
#import data from my github repositary file via csv
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/nadiyanaaz/World-education-pass-rate/main/pass_rate.csv')

df
```

```python
# display description
# of the dataset
df.info()
```

```python
#removing the dublicates from the data
df.drop_duplicates(inplace = True)

#droping the na coloumns
new_df = df.dropna()

print(df)
```

```python
# convert the state coloumn from object to string
df = df.astype({'State' : 'string'})
print(df.dtypes)
# Convert all columns to int dtype.
# This returns error in our DataFrame
#df = df.astype('int')
# convert "coloumns" from float to int and replace NaN values
df['Year'] = df['Year'].fillna(0).astype(int)
df['Num schools offering'] = df['Num schools offering'].fillna(0).astype(int)
df['Total Num'] = df['Total Num'].fillna(0).astype(int)
df['yield per teacher'] = df['yield per teacher'].fillna(0).astype(int)
df['diff from prev year'] = df['diff from prev year'].fillna(0).astype(int)
df['Pct increase'] = df['Pct increase'].fillna(0).astype(int)
df['Num passed'] = df['Num passed'].fillna(0).astype(int)
df['Pct passed'] = df['Pct passed'].fillna(0).astype(int)
df['Num female'] = df['Num female'].fillna(0).astype(int)
df['Pct female passed'] = df['Pct female passed'].fillna(0).astype(int)
df['Pct female'] = df['Pct female'].fillna(0).astype(int)
df['Num Black'] = df[' Num Black'].fillna(0).astype(int)
df['Num Black Passed'] = df['Num Black Passed'].fillna(0).astype(int)
df['Pct Black'] = df['Pct Black'].fillna(0).astype(int)
df['Num Hispanic'] = df['Num Hispanic'].fillna(0).astype(int)
df['Num Hispanic passed'] = df['Num Hispanic passed'].fillna(0).astype(str)
df['Pct Hispanic passed '] = df['Pct Hispanic passed'].fillna(0).astype(int)

print(df)
print(df.dtypes)
```

```python
#calculating the mean and mode of coloumns
df.mean()
df.mode()
df.mode()
df.dropna()
```

```python
#detecting outliers in data
# Create a boxplot of the 'Total Num' column
plt.boxplot(df['Total Num'], flierprops=red_circle)
red_circle = dict(markerfacecolor='red', marker='o')
plt.title('Total num')
plt.show()
```

```python
# Group the data by year and sum the 'Total Num' column for each year
num_students_per_year = df.groupby('Year')['Total Num'].sum()

# Plot the results in a histogram
plt.hist(num_students_per_year, bins=10, facecolor = 'yellow')
plt.xlabel('Number of Students')
plt.ylabel('Frequency')
plt.title('Number of Students per Year')
plt.show()
```
